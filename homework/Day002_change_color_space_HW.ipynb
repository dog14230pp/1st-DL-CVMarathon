```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('lena.png', cv2.IMREAD_COLOR)
imgH = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(imgH)
```

```python
imgL = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
plt.imshow(imgL)
```
